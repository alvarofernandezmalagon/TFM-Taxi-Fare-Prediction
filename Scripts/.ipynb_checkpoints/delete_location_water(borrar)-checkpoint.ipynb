# 4 Load image of Chicago map

### 4.1 Define the border and load the map of Chicago

```python
BB = (-87.95, -87.53, 41.66, 42.03)
chicago_map = plt.imread('../Data/maps/chicago_map.png')
```

### 4.2 Define the function to plot the Pickup locations y dropoff location in the map

```python
def plot_on_map(df, BB, chicago_map, s=10, alpha=0.2):
    fig, axs = plt.subplots(1, 2, figsize=(16,10))
    axs[0].scatter(df.pickup_centroid_longitude, df.pickup_centroid_latitude, zorder=1, alpha=alpha, c='b', s=s)
    axs[0].set_xlim((BB[0], BB[1]))
    axs[0].set_ylim((BB[2], BB[3]))
    axs[0].set_title('Pickup locations')
    axs[0].imshow(chicago_map, zorder=0, extent=BB)

    axs[1].scatter(df.dropoff_centroid_longitude, df.dropoff_centroid_latitude, zorder=1, alpha=alpha, c='r', s=s)
    axs[1].set_xlim((BB[0], BB[1]))
    axs[1].set_ylim((BB[2], BB[3]))
    axs[1].set_title('Dropoff locations')
    axs[1].imshow(chicago_map, zorder=0, extent=BB)
```

### 4.3 Plot the points

## 5 Check the number of points in water

### 5.1 Translate longitude/latitude coordinate into image xy coordinate

```python
#Define the function
def lonlat_to_xy(longitude, latitude, dx, dy, BB):
    return (dx*(longitude - BB[0])/(BB[1]-BB[0])).astype('int'),\
           (dy - dy*(latitude - BB[2])/(BB[3]-BB[2])).astype('int')
```

### 5.2 Define the threshold

### 5.2.1 Chek the number of points in water depending the value of the threshold

```python
water_points = {}
for threshold in np.arange(0,1,0.05):
    chicago_mask = plt.imread('../Data/maps/chicago_edited_map.png')[:,:,0] > threshold
    pickup_x, pickup_y = lonlat_to_xy(chicago_trips.pickup_centroid_longitude, chicago_trips.pickup_centroid_latitude, 
                                  chicago_mask.shape[1], chicago_mask.shape[0], BB)
    dropoff_x, dropoff_y = lonlat_to_xy(chicago_trips.dropoff_centroid_longitude, chicago_trips.dropoff_centroid_latitude, 
                                  chicago_mask.shape[1], chicago_mask.shape[0], BB)
    water = (chicago_mask[pickup_y, pickup_x] & chicago_mask[dropoff_y, dropoff_x])
    water_points[str(round(threshold,2))]=np.sum(water)
water_points
```

### 5.2.2 Plot the number of points in water depending the threshold

```python
# Unpack a list of pairs into two tuples
x, y = zip(*water_points.items()) 

#Plot the point in water per threshold
fig, ax = plt.subplots(2,1,figsize=(10,8))
plt.subplots_adjust(hspace=0.5)
fig.suptitle('POINTS IN WATER',fontsize=20)

#All points
ax[0].plot(x, y)
ax[0].set_title('Points in water per threshold',fontsize=16)
ax[0].set(xlabel='Threshold', ylabel='Points in Water')

# We zoom between 0.4 and 1 threshold because it is where the number of points in the water are stabilized.
ax[1].plot(x[8:], y[8:])
ax[1].set_title('Points in water threshold (0.4-1)',fontsize=16)
ax[1].set(xlabel='Threshold', ylabel='Points in Water')


plt.show()
```

We take the threshold 0.7 because the number of points in water are stabilized here

## 6 Remove points in water

### 6.1 Define the function to delete the points in water

```python
def remove_datapoints_from_water(df,BB,image,threshold=0.7):
    
    # read nyc mask and turn into boolean map with
    # land = True, water = False
    chicago_mask = plt.imread(image)[:,:,0] > threshold
    
    # calculate for each lon,lat coordinate the xy coordinate in the mask map
    pickup_x, pickup_y = lonlat_to_xy(df.pickup_centroid_longitude, df.pickup_centroid_latitude, 
                                      chicago_mask.shape[1], chicago_mask.shape[0], BB)
    dropoff_x, dropoff_y = lonlat_to_xy(df.dropoff_centroid_longitude, df.dropoff_centroid_latitude, 
                                      chicago_mask.shape[1], chicago_mask.shape[0], BB)    
    # calculate boolean index
    water = chicago_mask[pickup_y, pickup_x] & chicago_mask[dropoff_y, dropoff_x]
    
    # return only datapoints on land
    return df[~water]
```

### 6.2 Check the number of trips before delete 

```python
len(chicago_trips)
```

### 6.3 Remove the points in water

```python
chicago_trips = remove_datapoints_from_water(chicago_trips,
                                             BB,
                                             '../Data/maps/chicago_edited_map.png')
```
