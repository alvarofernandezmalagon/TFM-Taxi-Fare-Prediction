## 4 Join both Dataset

### 4.1 Define the keys to join both dataset

```python
# Taxi key
taxi_key=[taxi_trips['trip_start_timestamp'].dt.year,
          taxi_trips['trip_start_timestamp'].dt.month,
          taxi_trips['trip_start_timestamp'].dt.day,
          taxi_trips['trip_start_timestamp'].dt.hour]                                                                                          
```

```python
# Weather key
weather_key=[chicago_weather['datetime'].dt.year,
             chicago_weather['datetime'].dt.month,
             chicago_weather['datetime'].dt.day,
             chicago_weather['datetime'].dt.hour]
```

### 4.2 Join both Dataset

```python
taxi_trips = taxi_trips.merge(chicago_weather,
                               how='left',
                               left_on=taxi_key,
                               right_on=weather_key)
```

### estudiar los outliers

#### Obtenemos los campos de los cuales podemos obtener outliers

```python
df_ws_num=[]
for columna, tipo in zip(df_ws.dtypes.index,df_ws.dtypes.values):
    if tipo != "object":
        df_ws_num.append(columna)
df_ws_num
```

```python
df_ws_numerico = df_ws[df_ws_num].loc[:, 'time_in_hospital':'number_diagnoses']
df_ws_numerico.describe()
```

#### Almacenamos en un diccionario los medicamentos con sus threshold, q1 y q3

```python
thresholds={}
for columna in (df_ws_numerico.describe().dtypes.index):
    q1 = df_ws_numerico.describe().loc["25%"][columna]
    q3 = df_ws_numerico.describe().loc["75%"][columna]
    iqr = q3 - q1
    threshold = 1.5 * iqr
    thresholds[columna]=[threshold,q1,q3]
thresholds
```

#### Obtenemos el numeros de outliers con el porcentaje de outliers por variable

```python
outliers = {}
for i in thresholds:
    outlier = df_ws[df_ws_numerico[i] < (thresholds[i][1]-thresholds[i][0])].shape[0] \
    + df_ws[df_ws_numerico[i] > (thresholds[i][2]+thresholds[i][0])].shape[0]
    outliers[i] = [outlier,outlier/df_ws.shape[0]*100]
outliers
```

#### Graficamos

##### Outliers

```python
df_outliers = pd.DataFrame(outliers).T
df_outliers.columns = ['numero_outliers','%outliers']
df_outliers
```

```python
ax=df_outliers['%outliers'].plot(kind='bar', title ="Outliers", figsize=(15, 10), \
                                         fontsize=12, rot=45)
ax.set_xlabel("Campos", fontsize=12)
ax.set_ylabel("%outliers", fontsize=12)
plt.show()
```

### crear variables dummy en caso que sea conveniente

```python
def gender(campo):
    if campo=="Male":
        return 1
    elif campo=="Female":
        return 2
    else:
        return 0
```

```python
def age(campo):
    if campo in ["[0-10)","[10-20)","[20-30)"]:
        return 0
    elif campo in["[30-40)","[40-50)","[50-60)"]:
        return 1
    else:
        return 2
```

```python
def race(campo):
    if campo == "AfricanAmerican":
        return 0
    elif campo == "Asian":
        return 1
    elif campo == "Caucasian":
        return 2
    elif campo == "Hispanic":
        return 3
    else:
        return 4
```

```python
def change (campo):
    if campo == "No":
        return 0
    else:
        return 1
```

```python
def diabetesMed (campo):
    if campo == "No":
        return 0
    else:
        return 1
```

```python
def readmitted (campo):
    if campo == ">30":
        return 1
    elif campo == "<30":
        return 2
    else:
        return 0
```

```python
df_ws["gender_code"]=df_ws["gender"].apply(gender)
df_ws["age_code"]=df_ws["age"].apply(age)
df_ws["race_code"]=df_ws["race"].apply(race)
df_ws["change_code"] = df_ws["change"].apply(change)
df_ws["diabetesMed_code"] = df_ws["diabetesMed"].apply(diabetesMed)
df_ws["readmitted_code"] = df_ws["readmitted"].apply(readmitted)
df_ws.head()
```

# Exploratory Data Analysis

### crear la matriz de correlación

```python
df_ws_num=[]
for columna, tipo in zip(df_ws.dtypes.index,df_ws.dtypes.values):
    if tipo != "object":
        df_ws_num.append(columna)
df_ws_numerico = df_ws[df_ws_num].loc[:, 'time_in_hospital':]
corr = df_ws_numerico.corr()
corr
```

### pintar la matriz de correlación

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, annot= True, fmt= '.1f',cmap='PiYG', vmin= -1, vmax=1,center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

### determinar qué par de variables están correlacionadas

- El numero de medicamentos esta estrechamente relacionado con el cambio en los medicamentos para diabeticos y con la prescipcion de medicamentos
- Además el numero de medicaciones que toma la persona esta relacionado con el tiempo en el hospital y el numero de procesos.
- Existe tambien una relacion entre los cambios de medicaciones para diabeticos y si existe algun medicamento para la diabetes prescrito.
